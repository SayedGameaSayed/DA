### 📂 1. Data Loading and Initial Inspection

وبعدين في الـ Code Cell:



```python

# Load the dataset
df = pd.read_csv('jobs_in_data.csv')

# Display the first 5 rows
display(df.head())

# Check data types and non-null values
print("\nDataset Info:")
df.info()
```


4. تنظيف الداتا - Data Cleaning
### 🧹 2. Data Cleaning and Preprocessing
(خلية Markdown تشرح: هنشيل الصفوف المتكررة ونركز بس على المرتب بالدولار عشان نوحد المعايير)


```python
# Check for duplicates
print(f"Total duplicates before cleaning: {df.duplicated().sum()}")

# Drop duplicates
df_cleaned = df.drop_duplicates().copy()

# Select only relevant columns for our analysis
columns_to_keep = ['work_year', 'job_title', 'job_category', 'salary_in_usd', 
                   'experience_level', 'company_location']
df_cleaned = df_cleaned[columns_to_keep]

# Standardize experience level names for better readability
exp_map = {'EN': 'Entry-level', 'MI': 'Mid-level', 'SE': 'Senior', 'EX': 'Executive'}
if set(df_cleaned['experience_level'].unique()).intersection(set(exp_map.keys())):
    df_cleaned['experience_level'] = df_cleaned['experience_level'].replace(exp_map)

print(f"Dataset shape after cleaning: {df_cleaned.shape}")
```


5. التحليل والرسم - Exploratory Data Analysis (EDA)
هنا هنقسمها لـ 3 أجزاء، كل جزء بيجاوب على سؤال من اللي حددناهم فوق.

#### Q1: What are the top-paying job titles? (Markdown)



```python
# Top 10 jobs by average salary
top_jobs = df_cleaned.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=top_jobs.values, y=top_jobs.index, color='#7b8cd1')

plt.title('Top 10 Highest Paying Data Roles', fontsize=14, pad=15)
plt.xlabel('Average Salary (USD)', fontsize=11)
plt.ylabel('')

# Clean up borders (Minimalist look)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```


#### Q2: How does experience level affect salary? (Markdown)


```python
# Define the logical order of experience
experience_order = ['Entry-level', 'Mid-level', 'Senior', 'Executive']
existing_levels = [level for level in experience_order if level in df_cleaned['experience_level'].unique()]

plt.figure(figsize=(9, 5))
ax = sns.barplot(data=df_cleaned, x='experience_level', y='salary_in_usd', 
                 order=existing_levels, color='#7b8cd1', errorbar=None)

plt.title('Average Salary by Experience Level', fontsize=14, pad=15)
plt.xlabel('Experience Level', fontsize=11)
plt.ylabel('Average Salary (USD)', fontsize=11)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```


#### Q3: Which countries offer the highest average salaries? (Markdown)

```python


# Filter countries with at least 10 data points to ensure statistical significance
country_counts = df_cleaned['company_location'].value_counts()
valid_countries = country_counts[country_counts > 10].index

top_countries = df_cleaned[df_cleaned['company_location'].isin(valid_countries)]
top_countries = top_countries.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
ax = sns.barplot(x=top_countries.index, y=top_countries.values, color='#7b8cd1')

plt.title('Top 10 Countries with Highest Salaries (Min. 10 records)', fontsize=14, pad=15)
plt.xlabel('Country', fontsize=11)
plt.ylabel('Average Salary (USD)', fontsize=11)
plt.xticks(rotation=45)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
